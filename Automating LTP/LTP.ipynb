```python
#rEMOVING COMPANIES WITH CLOSE LESS THAN 9
LTD = LTD[LTD[' CLOSE_PRICE']>=9]
```

```python
#CONSIDERING TOTAL TRADED QUANTITY TO BE MORE THAN 99999
LTD = LTD[LTD[' TTL_TRD_QNTY']>99999]
```

```python
#CALCULATING PERCENTAGE CHANGE IN LTP VS CP PER CP OF STOCK = 100*(CP-LTP)/CP
#We want this to be negative i.e last trade price should be more than close price.

LTD[' LAST_PRICE'] = pd.to_numeric(LTD[' LAST_PRICE'], errors='coerce')
LTD[' CLOSE_PRICE'] = pd.to_numeric(LTD[' CLOSE_PRICE'], errors='coerce')
LTD['%ltp_vs_cp'] = 100*(LTD[' CLOSE_PRICE']-LTD[' LAST_PRICE'])/LTD[' CLOSE_PRICE']
LTD = LTD[LTD['%ltp_vs_cp']<0]
```

```python
#CALCULATING DIFFERENCE PERCENTAGE CHANGE IN PREV_CLOSE_PRICE AND CLOSE_PRICE WRT PREV_CLOSE_PRICE
#We are calculating indirectly daily return percentage for last day which should be positive.

LTD['%close_change'] = 100*(LTD[' CLOSE_PRICE']-LTD[' PREV_CLOSE'])/LTD[' PREV_CLOSE']
LTD = LTD[LTD['%close_change']>0]
```

```python
#CALCULATING DIFFERENCE PERCENTAGE CHANGE IN OPEN_PRICE AND CLOSE_PRICE WRT OPEN_PRICE
LTD['%open_close'] = 100*(LTD[' CLOSE_PRICE']-LTD[' OPEN_PRICE'])/LTD[' OPEN_PRICE']
LTD = LTD[LTD['%open_close']>0]
```

```python
#Filtering stocks with less than (-33.33%) of above value.

LTD['WeightVs%chg'] = 100*LTD['%ltp_vs_cp']/LTD['%open_close']
LTD = LTD[LTD['WeightVs%chg']<(-1)*33.33]
```

```python
LTD['WeightVs%Dchg'] = 100*LTD['%ltp_vs_cp']/LTD['%close_change']
LTD = LTD[LTD['WeightVs%Dchg']<(-1)*33.33]
```

```python
# Sort the DataFrame by symbol and date


# Filter stocks with bullish engulfing pattern
filtered_stocks = LTD2.groupby('SYMBOL').apply(lambda x: (x['script'] == 1) &
                                                       (x['BODY%'] > 65 ) &
                                                       (x['Prchng']>=2.5) &
                                                       (x['Prchng'].shift()>=2.5) &
                                                       (x['Prchng'].shift(2)>=2.5) &
                                                       ((x[' OPEN_PRICE'].shift(2)-x[' CLOSE_PRICE'].shift(2))<0) &
                                                       ((x[' OPEN_PRICE'].shift()-x[' CLOSE_PRICE'].shift())<0) &
                                                       ((x[' OPEN_PRICE']-x[' CLOSE_PRICE'])<0)
                                                       )


# Reset index to remove the groupby index
filtered_stocks = filtered_stocks.reset_index(level=0, drop=True)

# Get the names of stocks with bullish engulfing pattern
stock_names = filtered_stocks[filtered_stocks].index.unique()

# Print the stock names
# Get the rows based on the indexes
bullen = LTD2.loc[stock_names]
bullen['TYPE'] = 'threewhitesoldiers'
bullen.to_csv("threewhitesoldiers.csv",index=False)
bullen
```

### GAP DOWN BULLISH

```python
# Sort the DataFrame by symbol and date

LTD3['Candle Color'] = LTD.apply(lambda row: 'Green' if row[' CLOSE_PRICE'] > row[' OPEN_PRICE'] else 'Red', axis=1)
LTD3['% Gap Change'] = (LTD3[' OPEN_PRICE'] - LTD3[' CLOSE_PRICE'].shift(1))*100/LTD3[' CLOSE_PRICE'].shift(1)
LTD3['Body Percent'] = ((LTD3[' CLOSE_PRICE'] - LTD3[' OPEN_PRICE']) / (LTD3[' HIGH_PRICE'] - LTD3[' LOW_PRICE'])) * 100

# Filter stocks with bullish engulfing pattern
filtered_stocks = LTD3.groupby('SYMBOL').apply(lambda x: (x['script'] == 1) &
                                              (x['Body Percent'] > 85) &  # Condition 1: Body Percent > 85%
                                              (x['Candle Color'] == 'Green') &  # Condition 2: Candle is green
                                              (x['% Gap Change'] < -3)
                                              )

# Reset index to remove the groupby index
filtered_stocks = filtered_stocks.reset_index(level=0, drop=True)

# Get the names of stocks with Gap down bullish
stock_names = filtered_stocks[filtered_stocks].index.unique()

# Print the stock names
# Get the rows based on the indexes
gapdown = LTD3.loc[stock_names]
gapdown['TYPE'] = 'gapdown bull'
gapdown.to_csv("gapdownbullish.csv",index=False)
gapdown
```

```python
def compare_open_low(group):
    # Check if the script is 1
    if (group['script'] == 1).any():
        # Get the open price value for script 1
        open_price_script_1 = group.loc[group['script'] == 1, ' OPEN_PRICE'].values[0]
        # Check if the script is 0
        if (group['script'] == 0).any():
            # Get the low price value for script 0
            low_price_script_0 = group.loc[group['script'] == 0, ' LOW_PRICE'].values[0]
            # Compare open_price_script_1 with low_price_script_0 and assign the result
            group['Lowcheck'] = (group['script'] == 1) & (open_price_script_1 > low_price_script_0)
        else:
            # No script 0, assign 0 to all rows
            group['Lowcheck'] = 0
    else:
        # No script 1, assign 0 to all rows
        group['Lowcheck'] = 0
    return group

LTD3['Lowcheck'] = LTD3.groupby('SYMBOL').apply(compare_open_low)['Lowcheck']
```

```python
def compare_close_high(group):
    # Check if the script is 1
    if (group['script'] == 1).any():
        # Get the close price value for script 1
        close_price_script_1 = group.loc[group['script'] == 1, ' CLOSE_PRICE'].values[0]
        # Check if the script is 0
        if (group['script'] == 0).any():
            # Get the high price value for script 0
            high_price_script_0 = group.loc[group['script'] == 0, ' HIGH_PRICE'].values[0]
            # Compare close_price_script_1 with high_price_script_0 and assign the result
            group['highcheck'] = (group['script'] == 1) & (close_price_script_1 > high_price_script_0)
        else:
            # No script 0, assign 0 to all rows
            group['highcheck'] = 0
    else:
        # No script 1, assign 0 to all rows
        group['highcheck'] = 0
    return group

LTD3['highcheck'] = LTD3.groupby('SYMBOL').apply(compare_close_high)['highcheck']
```

```python
def compare_prchng_zero(group):
    # Check if the script is 0
    if (group['script'] == 0).any():
        # Check if Prchng is less than zero for script 0
        prchng_script_0 = group.loc[group['script'] == 0, 'Prchng'].values[0]
        group['prevClose'] = (group['script'] == 1) & (prchng_script_0 < 0)
    else:
        # No script 0, assign 0 to all rows
        group['prevClose'] = 0
    return group

LTD3['prevClose'] = LTD3.groupby('SYMBOL').apply(compare_prchng_zero)['prevClose']
```

```python
# List of CSV files to combine
csv_files = ["bullengulfing.csv", "bearengulfing.csv", "bullishnews.csv", "bearishnews.csv", "ltp_cp.csv", "threewhitesoldiers.csv", "threeblackcrows.csv", "W.csv", "M.csv","bullpairedcandle.csv","bearpairedcandle.csv","mbullishbreakout.csv","mbearishbreakout.csv","qbullishbreakout.csv","qbearishbreakout.csv","ybullishbreakout.csv","ybearishbreakout.csv","gapdownbullish.csv"]

# Output file to save the combined data
output_file = "combined_data.csv"

# Desired column names to keep in the final CSV
columns_to_keep = [" DATE1", "TYPE", " SERIES", "SYMBOL", " CLOSE_PRICE", "color", "Prchng"," LOW_PRICE"," HIGH_PRICE"]

# Create a set to store all unique column names from all CSV files
all_columns = set()

# Loop through each CSV file and collect unique column names
for file in csv_files:
    data = pd.read_csv(file, nrows=0)
    all_columns.update(data.columns)

# Find the common columns present in all CSV files and keep only the desired columns
common_columns = list(all_columns.intersection(columns_to_keep))

# Create an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Loop through each CSV file and concatenate them into the combined_data DataFrame
for file in csv_files:
    data = pd.read_csv(file, usecols=common_columns)
    combined_data = combined_data.append(data, ignore_index=True)

# Save the combined data to the output_file
combined_data.to_csv(output_file, index=False)
print(f"Combined data with specific columns saved to {output_file}")
```
